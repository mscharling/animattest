````python
from manim import *
````

````python
%%manim -v WARNING -qm -t --format=gif SquareToCircle

class SquareToCircle(Scene):
   def construct(self):
      square = Square()
      square.set_color(GREEN)
      circle = Circle()
      circle.set_fill(PINK, opacity=0.5)
      self.play(Create(square))
      self.play(Transform(square, circle))
      self.wait()
````

En rod bestr af en *radikand* og en *rodeksponent*. $\sqrt[n]{\ }$ kaldes den $n$'te rod. Eftersom rdder er det modsatte af potenser, s kan vi formulere, hvad vi mener med en rod ved hjlp af potenser:

```{prf:definition} Rdder
:label: roedder

Lad $a$ vre et vilkrligt tal. Hvis $b$ opfylder

$$

    \sqrt[n]{a} = b \ \text{nr}\ b^n = a,

$$

s kalder vi $b$ den $n$'te rod af $a$. 

```
